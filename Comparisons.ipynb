```python
import urllib.request
```

```python
# Reading sudoku from the provided source
def readSudoku(url):

  # Get the whole content of the source
  with urllib.request.urlopen(url) as response:
    content = response.read().decode("utf-8")

  # Filter out rows (a line is a row)
  rows = [line for line in content.split('\r\n')]

  # Convert each row in a list of integers
  # Getting all the elemets on Sudoku grid
  grid = [[],[],[],[],[],[],[],[],[]]
  for row in range(0,9):
    grid[row] = [int(x) for x in str(rows[row])]

  return grid
```

```python
def print_grid(grid):
  for row in range(9):
    if row % 3 == 0 and row != 0:
      print("- - - - - - - - - - - - - - ")
    for col in range(9):
      if col % 3 == 0 and col != 0:
        print(" | ", end="")
      if col == 8:
        print(grid[row][col])
      else:
        print(f'{str(grid[row][col])} ', end="")
```

```python
grid = readSudoku("https://uchida.cis.k.hosei.ac.jp/2015/optBprob-easy.txt")
print_grid(grid)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(n):
    return 1.3**n

# Generate x values from 0 to 10
x = np.arange(0, 11)

# Calculate y values using the function
y = f(x)

# Plot the function
plt.plot(x, y, 'b-')
plt.xlabel('n')
plt.ylabel('O(1.3^n)')
plt.title('Plot of O(1.3^n)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(1, 10, 100)

# Calculate y values for O(1.3^n)
y1 = 1.3 ** x

# Calculate y values for O(n^2)
y2 = x ** 2

# Plot the functions
plt.plot(x, y1, label='O(1.3^n)')
plt.plot(x, y2, label='O(n^2)')

# Add labels and title
plt.xlabel('n')
plt.ylabel('Runtime')
plt.title('O(1.3^n) vs O(n^2)')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for n
n = np.arange(1, 10)

# Calculate the corresponding values for O(1.3^(n^2)) and O(n^2)
o1 = 1.3 ** (n ** 2)
o2 = n ** 2

# Plot the functions
plt.plot(n, o1, label='O(1.3^(n^2))')
plt.plot(n, o2, label='O(n^2)')

# Add labels and title to the plot
plt.xlabel('n')
plt.ylabel('Time Complexity')
plt.title('O(1.3^(n^2)) vs O(n^2)')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
